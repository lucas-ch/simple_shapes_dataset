```python
dataset = generate_dataset_biased(
    n_samples= 9,
    class_configs = {0: {'bias_rate': 1.0, 'fixed_color_hls': np.array([0,128,255])}, 
                     1: {'bias_rate': 1.0, 'fixed_color_hls': np.array([60,128,255])},
                     2: {'bias_rate': 1.0, 'fixed_color_hls': np.array([120,128,255])}},
    min_scale=7,
    max_scale=14,
    min_lightness=46,
    max_lightness=256,
    imsize=32)
```

```python
%matplotlib inline

from typing import cast
from matplotlib import pyplot as plt
from matplotlib.axes import Axes

dpi = 1
imsize = 32
enumerator = tqdm(
        enumerate(
            zip(
                dataset.classes,
                dataset.locations,
                dataset.sizes,
                dataset.rotations,
                dataset.colors,
                strict=True,
            )
        ),
        total=len(dataset.classes),
    )

for k, (cls, location, size, rotation, color) in enumerator:
    # Création de la figure
    fig, ax = plt.subplots(figsize=(imsize / dpi, imsize / dpi), dpi=dpi)
    ax = cast(plt.Axes, ax)
    generate_image(ax, cls, location, size, rotation, color, imsize)
    ax.set_facecolor("black")    
    plt.show()
    plt.close(fig)
```
